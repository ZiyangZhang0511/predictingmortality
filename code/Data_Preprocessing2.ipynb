```python
## add Creatinine values to Features

features[['Creatinine Minimum', 'Creatinine Maximum', 'Creatinine Max during day2 and day3']] = 0

table = chartevents_creatinine[['stay_id', 'charttime', 'value']]
table['charttime'] = pd.to_datetime(table['charttime'])

target_3 = pd.merge(target_2, table, how='left', on='stay_id')

target_3['the_first_day'] = target_3['intime'] + pd.Timedelta(days=1)
target_3['the_third_day'] = target_3['intime'] + pd.Timedelta(days=3)

stay_id = target_3['stay_id'].unique()

for i in stay_id:
    t0 = target_3[target_3['stay_id'] == i]
    day1 = t0['the_first_day'].sample(n=1).iloc[0]
    day3 = t0['the_third_day'].sample(n=1).iloc[0]

    # get day1 extremum
    t1 = t0[t0['charttime'] < day1]
    values_day1 = t1['value'].values.astype(float)
    if values_day1.size > 0:
        values_day1_max = np.max(values_day1)
        values_day1_min = np.min(values_day1)
    else:
        values_day1_max = np.nan
        values_day1_min = np.nan
    features.loc[features['stay_id']==i, 'Creatinine Minimum'] = values_day1_min
    features.loc[features['stay_id']==i, 'Creatinine Maximum'] = values_day1_max
    # display(features)

    # get day2today3 maximum
    t2 = t0[(t0['charttime'] < day3) & (t0['charttime'] > day1)]
    values_day2today3 = t2['value'].values.astype(float)
    if values_day2today3.size > 0:
        values_day2today3_max = np.max(values_day2today3)
    else:
        values_day2today3_max = np.nan
    features.loc[features['stay_id']==i, 'Creatinine Max during day2 and day3'] = values_day2today3_max
```

```python
chartevents_gcs

def calculate_gcs_motor(valuenum, itemid):
    return valuenum if itemid == 223901 else None

def calculate_gcs_verbal(valuenum, value, itemid):
    if itemid == 223900:
        if value == 'No Response-ETT':
            return 0
        else:
            return valuenum
    else:
        return None

def calculate_gcs_eyes(valuenum, itemid):
    return valuenum if itemid == 220739 else None

def calculate_endotrachflag(value, itemid):
    if itemid == 223900 and value == 'No Response-ETT':
        return 1
    else:
        return 0

# Apply the functions to create new columns
chartevents_gcs['GCSMotor'] = chartevents_gcs.apply(lambda x: calculate_gcs_motor(x['valuenum'], x['itemid']), axis=1)
chartevents_gcs['GCSVerbal'] = chartevents_gcs.apply(lambda x: calculate_gcs_verbal(x['valuenum'], x['value'], x['itemid']), axis=1)
chartevents_gcs['GCSEyes'] = chartevents_gcs.apply(lambda x: calculate_gcs_eyes(x['valuenum'], x['itemid']), axis=1)
chartevents_gcs['endotrachflag'] = chartevents_gcs.apply(lambda x: calculate_endotrachflag(x['value'], x['itemid']), axis=1)

# Group by subject_id, stay_id, charttime and aggregate
aggregated_df = chartevents_gcs.groupby(['subject_id', 'stay_id', 'charttime']).agg(
    GCSMotor=('GCSMotor', 'max'),
    GCSVerbal=('GCSVerbal', 'max'),
    GCSEyes=('GCSEyes', 'max'),
    endotrachflag=('endotrachflag', 'max')
).reset_index()

# Generate row numbers partitioned by stay_id and ordered by charttime
aggregated_df['rn'] = aggregated_df.sort_values(by=['charttime']).groupby(['stay_id']).cumcount() + 1
```

```python
df_base = aggregated_df
df_base['charttime'] = pd.to_datetime(df_base['charttime'])

# Create a shifted DataFrame to simulate the self-join for getting previous values
df_base_shifted = df_base.copy()
df_base_shifted['rn'] += 1  # Shift the row number to join on the next row in the sequence

# Merge the original DataFrame with the shifted one
df_joined = pd.merge(df_base, df_base_shifted, how='left', left_on=['stay_id', 'rn'], right_on=['stay_id', 'rn'], suffixes=('', '_Prev'))

# Filter out rows where the charttime difference is more than 6 hours
df_joined = df_joined[abs(df_joined['charttime'] - df_joined['charttime_Prev']) <= pd.Timedelta(hours=6)]
```

```python
df_joined['stay_id'].nunique()
```

```python
def calculate_gcs(row):
    # If current verbal is 0 or null with previous 0, indicating sedation, GCS is 15
    if row['GCSVerbal'] == 0 or (pd.isnull(row['GCSVerbal']) and row['GCSVerbal_Prev'] == 0):
        return 15
    # If the previous verbal score indicates intubation, and there's no current score, use default values
    elif row['GCSVerbal_Prev'] == 0:
        return coalesce(row['GCSMotor'], 6) + coalesce(row['GCSVerbal'], 5) + coalesce(row['GCSEyes'], 4)
    # Otherwise, use current or previous values, with defaults if neither is available
    else:
        return (coalesce(row['GCSMotor'], coalesce(row['GCSMotor_Prev'], 6)) +
                coalesce(row['GCSVerbal'], coalesce(row['GCSVerbal_Prev'], 5)) +
                coalesce(row['GCSEyes'], coalesce(row['GCSEyes_Prev'], 4)))

# Helper function to coalesce values
def coalesce(*args):
    for value in args:
        if not pd.isnull(value):
            return value
    return None

# Apply the GCS calculation to each row
df_joined['GCS'] = df_joined.apply(calculate_gcs, axis=1)
```

```python
df_joined['GCS'].describe()
```

```python
df_gcs_stg = df_joined.copy()

def calculate_components_measured(row):
    # Using the Python equivalent of COALESCE() by using the 'or' operator to return the first non-null value
    components_measured = 0

    # Checking each pair and incrementing components_measured if the coalesce of each pair is not null
    components_measured += 1 if (row['GCSMotor'] or row['GCSMotor_Prev']) is not None else 0
    components_measured += 1 if (row['GCSVerbal'] or row['GCSVerbal_Prev']) is not None else 0
    components_measured += 1 if (row['GCSEyes'] or row['GCSEyes_Prev']) is not None else 0

    return components_measured
# Helper function to coalesce values
def coalesce(*args):
    for value in args:
        if not pd.isnull(value):
            return value
    return None

df_gcs_stg['GCSMotor'] = df_gcs_stg['GCSMotor'].fillna(df_gcs_stg['GCSMotor_Prev'])
df_gcs_stg['GCSVerbal'] = df_gcs_stg['GCSVerbal'].fillna(df_gcs_stg['GCSVerbal_Prev'])
df_gcs_stg['GCSEyes'] = df_gcs_stg['GCSEyes'].fillna(df_gcs_stg['GCSEyes_Prev'])

df_gcs_stg['components_measured'] = df_gcs_stg.apply(calculate_components_measured, axis=1)
```

```python
final_df_list = []

# Define chunk size
chunk_size = 100000  # Adjust based on your system's memory capacity

for start in range(0, stg2.shape[0], chunk_size):
    end = min(start + chunk_size, stg2.shape[0])
    stg2_chunk = stg2.iloc[start:end]

    # Merge operation with stg_spo2
    merged_chunk = pd.merge(stg2_chunk, stg_fio2, on='subject_id', how='left', suffixes=('', '_s2'))

    # Apply filters
    merged_chunk = merged_chunk[(merged_chunk['charttime_s2'] >= (merged_chunk['charttime'] - pd.Timedelta(hours=4))) &
                                (merged_chunk['charttime_s2'] <= merged_chunk['charttime']) &
                                (merged_chunk['fio2_chartevents'] > 0)]

    # Ensure we're only working with the rows with the most recent SpO2
    merged_chunk = merged_chunk[merged_chunk['lastRowSpO2'] == 1]

    # Sort the chunk
    merged_chunk.sort_values(by=['subject_id', 'charttime', 'charttime_s2'], ascending=[True, True, False], inplace=True)
    merged_chunk['lastRowFiO2'] = merged_chunk.groupby(['subject_id', 'charttime']).cumcount() + 1

    # Calculate specimen_prob
    def calculate_specimen_prob(row):
        specimen_prob = 1 / (1 + np.exp(-(-0.02544
                                      + 0.04598 * row['po2']
                                      + (-0.15356 * row['SpO2'] if pd.notnull(row['SpO2']) else -0.15356 * 97.49420 + 0.13429)
                                      + (0.00621 * row['fio2_chartevents'] if pd.notnull(row['fio2_chartevents']) else 0.00621 * 51.49550 - 0.24958)
                                      + (0.10559 * row['hemoglobin'] if pd.notnull(row['hemoglobin']) else 0.10559 * 10.32307 + 0.05954)
                                      + (0.13251 * row['so2'] if pd.notnull(row['so2']) else 0.13251 * 93.66539 - 0.23172)
                                      + (-0.01511 * row['pco2'] if pd.notnull(row['pco2']) else -0.01511 * 42.08866 - 0.01630)
                                      + (0.01480 * row['fio2'] if pd.notnull(row['fio2']) else 0.01480 * 63.97836 - 0.31142)
                                      + (-0.00200 * row['aado2'] if pd.notnull(row['aado2']) else -0.00200 * 442.21186 - 0.01328)
                                      + (-0.03220 * row['bicarbonate'] if pd.notnull(row['bicarbonate']) else -0.03220 * 22.96894 - 0.06535)
                                      + (0.05384 * row['totalco2'] if pd.notnull(row['totalco2']) else 0.05384 * 24.72632 - 0.01405)
                                      + (0.08202 * row['lactate'] if pd.notnull(row['lactate']) else 0.08202 * 3.06436 + 0.06038)
                                      + (0.10956 * row['ph'] if pd.notnull(row['ph']) else 0.10956 * 7.36233 - 0.00617)
                                      + (0.00848 * row['o2flow'] if pd.notnull(row['o2flow']) else 0.00848 * 7.59362 - 0.35803)
                                     )))
        return specimen_prob

    merged_chunk['specimen_prob'] = merged_chunk.apply(calculate_specimen_prob, axis=1)

    # Filter to keep only the most recent SpO2 measurement for each blood gas record within the chunk
    # final_chunk = merged_chunk[merged_chunk['lastRowSpO2'] == 1]
    final_chunk = merged_chunk

    # Append the processed chunk to the list
    final_df_list.append(final_chunk)

# Combine all processed chunks into a final DataFrame
final_df = pd.concat(final_df_list, ignore_index=True)
stg3 = final_df
```

```python
merged = pd.merge(tm, ventilator_setting, on=['stay_id', 'charttime'], how='left', suffixes=('', '_vs'))
merged = pd.merge(merged, oxygen_delivery, on=['stay_id', 'charttime'], how='left', suffixes=('', '_od'))

# Define a function to categorize ventilation status
def categorize_ventilation_status(row):
    o2_device_1 = row['o2_delivery_device_1']
    ventilator_mode = row['ventilator_mode']
    ventilator_mode_hamilton = row['ventilator_mode_hamilton']

    # Define the categories based on o2 device and ventilation mode
    trach_devices = ['Tracheostomy tube']
    conditions1 = ['(S) CMV', 'APRV', 'APRV/Biphasic+ApnPress', 'APRV/Biphasic+ApnVol',
        'APV (cmv)', 'Ambient', 'Apnea Ventilation', 'CMV', 'CMV/ASSIST', 'CMV/ASSIST/AutoFlow',
        'CMV/AutoFlow', 'CPAP/PPS', 'CPAP/PSV+Apn TCPL', 'CPAP/PSV+ApnPres', 'CPAP/PSV+ApnVol',
        'MMV', 'MMV/AutoFlow', 'MMV/PSV', 'MMV/PSV/AutoFlow', 'P-CMV', 'PCV+', 'PCV+/PSV',
        'PCV+Assist', 'PRES/AC', 'PRVC/AC', 'PRVC/SIMV', 'PSV/SBT', 'SIMV', 'SIMV/AutoFlow',
        'SIMV/PRES', 'SIMV/PSV', 'SIMV/PSV/AutoFlow', 'SIMV/VOL', 'SYNCHRON MASTER',
        'SYNCHRON SLAVE', 'VOL/AC']

    conditions2 = ['APRV', 'APV (cmv)', 'Ambient', '(S) CMV',
        'P-CMV', 'SIMV', 'APV (simv)', 'P-SIMV', 'VS', 'ASV']

    conditions3 = ['DuoPaP', 'NIV', 'NIV-ST']

    conditions4 = ['Nasal cannula', 'Face tent', 'Aerosol-cool', 'Non-rebreather',
        'Venti mask ', 'Medium conc mask ', 'T-piece', 'Ultrasonic neb',
        'Vapomist', 'Oxymizer']

    if o2_device_1 in trach_devices:
        return 'Trach'
    elif o2_device_1 in ['Endotracheal tube'] or ventilator_mode in conditions1 or ventilator_mode_hamilton in conditions2:
        return 'InvasiveVent'
    elif o2_device_1 in ['Bipap mask ', 'CPAP mask '] or ventilator_mode_hamilton in conditions3:
        return 'NonInvasiveVent'
    elif o2_device_1 in ['High flow neb', 'High flow nasal cannula']:
        return 'HighFlow'
    elif o2_device_1 in conditions4:
        return 'Oxygen'
    else:
        return None

merged['vent_mode'] = merged['ventilator_mode'].fillna(merged['ventilator_mode_hamilton'])

# Apply the categorization function
merged['ventilation_status'] = merged.apply(categorize_ventilation_status, axis=1)

# Select the relevant columns
merged = merged[['subject_id', 'stay_id', 'charttime', 'o2_delivery_device_1', 'vent_mode', 'ventilation_status']]
```

```python
vs = merged.copy()
```

```python
vs_filtered = vs[vs['ventilation_status'].notnull()]

# Sort by stay_id and charttime for accurate shift operations
vs_sorted = vs_filtered.sort_values(by=['stay_id', 'charttime'])

# Group by stay_id and ventilation_status to calculate charttime_lag
vs_sorted['charttime_lag'] = vs_sorted.groupby(['stay_id', 'ventilation_status'])['charttime'].shift(1)

# Group by stay_id to calculate charttime_lead and ventilation_status_lag
vs_sorted['charttime_lead'] = vs_sorted.groupby(['stay_id'])['charttime'].shift(-1)
vs_sorted['ventilation_status_lag'] = vs_sorted.groupby(['stay_id'])['ventilation_status'].shift(1)

# Select relevant columns for vd0
vd0 = vs_sorted[['stay_id', 'charttime', 'o2_delivery_device_1', 'vent_mode',
                 'charttime_lag', 'charttime_lead', 'ventilation_status', 'ventilation_status_lag']]
```

```python
vd0.info()
```

```python
vd0['charttime'] = pd.to_datetime(vd0['charttime'])
vd0['charttime_lag'] = pd.to_datetime(vd0['charttime_lag'])
vd0['charttime_lead'] = pd.to_datetime(vd0['charttime_lead'])

# Calculate the time difference for ventduration
vd0['ventduration'] = (vd0['charttime'] - vd0['charttime_lag']).dt.total_seconds() / 3600

# Determine if the current ventilation status is new
vd0['new_status'] = (
    ((vd0['charttime'] - vd0['charttime_lag']).dt.total_seconds() / 3600 >= 14) |
    (vd0['ventilation_status_lag'].isnull()) |
    (vd0['ventilation_status_lag'] != vd0['ventilation_status'])
).astype(int)

# Select relevant columns for vd1
vd1 = vd0[['stay_id', 'o2_delivery_device_1', 'vent_mode', 'charttime_lag', 'charttime', 'charttime_lead',
           'ventilation_status', 'ventduration', 'new_status']]
```

```python
vd1.info()
```

```python
vd1['charttime'] = pd.to_datetime(vd1['charttime'])

# Sort by stay_id and charttime to ensure correct order for cumsum
vd1_sorted = vd1.sort_values(by=['stay_id', 'charttime'])

# Calculate the cumulative sum of new_status within each stay_id
vd1_sorted['vent_num'] = vd1_sorted.groupby('stay_id')['new_status'].cumsum()

# The resulting DataFrame vd1_sorted now corresponds to vd2 in the SQL query
vd2 = vd1_sorted
```

```python
vd2.info()
```

```python
vd2['charttime'] = pd.to_datetime(vd2['charttime'])
vd2['charttime_lead'] = pd.to_datetime(vd2['charttime_lead'])

# Calculate endtime based on conditions
def calculate_endtime(row):
    if pd.isnull(row['charttime_lead']) or ((row['charttime_lead'] - row['charttime']).total_seconds() / 3600 >= 14):
        return row['charttime']
    else:
        return row['charttime_lead']

vd2['endtime'] = vd2.apply(calculate_endtime, axis=1)

# Group by stay_id and vent_num to aggregate data
aggregated = vd2.groupby(['stay_id', 'vent_num']).agg(
    starttime=('charttime', 'min'),
    endtime=('endtime', 'max'),
    ventilation_status=('ventilation_status', 'max')  # Assuming ventilation_status is consistent within each group
).reset_index()

# Filter out groups where start and end times are the same
aggregated = aggregated[aggregated['starttime'] != aggregated['endtime']]
```

## score calculation

```python
def respiration_score(row):
    if row['PaO2FiO2_vent_min'] < 100: return 4
    elif row['PaO2FiO2_vent_min'] < 200: return 3
    elif row['PaO2FiO2_novent_min'] < 300: return 2
    elif row['PaO2FiO2_novent_min'] < 400: return 1
    elif pd.isnull(row['PaO2FiO2_vent_min']) and pd.isnull(row['PaO2FiO2_novent_min']):
        return np.nan
    else: return 0

def coagulation_score(row):
    if row['platelet_min'] < 20: return 4
    elif row['platelet_min'] < 50: return 3
    elif row['platelet_min'] < 100: return 2
    elif row['platelet_min'] < 150: return 1
    elif pd.isnull(row['platelet_min']): return np.nan
    else: return 0

def liver_score(row):
    if row['bilirubin_max'] >= 12.0: return 4
    elif row['bilirubin_max'] >= 6.0: return 3
    elif row['bilirubin_max'] >= 2.0: return 2
    elif row['bilirubin_max'] >= 1.2: return 1
    elif pd.isnull(row['bilirubin_max']): return np.nan
    else: return 0

def cardiovascular_score(row):
    if (row['rate_dopamine'] > 15) or (row['rate_epinephrine'] > 0.1) or (row['rate_norepinephrine'] > 0.1):
        return 4
    elif (row['rate_dopamine'] > 5) or (row['rate_epinephrine'] <= 0.1 and row['rate_epinephrine'] > 0) or \
         (row['rate_norepinephrine'] <= 0.1 and row['rate_norepinephrine'] > 0):
        return 3
    elif (row['rate_dopamine'] > 0) or (row['rate_dobutamine'] > 0):
        return 2
    elif row['mbp_min'] < 70:
        return 1
    elif pd.isnull(row['mbp_min']) and pd.isnull(row['rate_dopamine']) and pd.isnull(row['rate_dobutamine']) and \
         pd.isnull(row['rate_epinephrine']) and pd.isnull(row['rate_norepinephrine']):
        return np.nan
    else:
        return 0

def cns_score(row):
    if row['gcs_min'] >= 13 and row['gcs_min'] <= 14:
        return 1
    elif row['gcs_min'] >= 10 and row['gcs_min'] <= 12:
        return 2
    elif row['gcs_min'] >= 6 and row['gcs_min'] <= 9:
        return 3
    elif row['gcs_min'] < 6:
        return 4
    elif pd.isnull(row['gcs_min']):
        return np.nan
    else:
        return 0

def renal_score(row):
    if row['creatinine_max'] >= 5.0:
        return 4
    elif row['UrineOutput'] < 200:
        return 4
    elif row['creatinine_max'] >= 3.5 and row['creatinine_max'] < 5.0:
        return 3
    elif row['UrineOutput'] < 500:
        return 3
    elif row['creatinine_max'] >= 2.0 and row['creatinine_max'] < 3.5:
        return 2
    elif row['creatinine_max'] >= 1.2 and row['creatinine_max'] < 2.0:
        return 1
    elif pd.isnull(row['UrineOutput']) and pd.isnull(row['creatinine_max']):
        return np.nan
    else: return 0

scorecomp['respiration'] = scorecomp.apply(respiration_score, axis=1)
scorecomp['coagulation'] = scorecomp.apply(coagulation_score, axis=1)
scorecomp['liver'] = scorecomp.apply(liver_score, axis=1)
scorecomp['cardiovascular'] = scorecomp.apply(cardiovascular_score, axis=1)
scorecomp['cns'] = scorecomp.apply(cns_score, axis=1)
scorecomp['renal'] = scorecomp.apply(renal_score, axis=1)

scorecalc = scorecomp[['stay_id', 'respiration', 'coagulation', 'liver', 'cardiovascular', 'cns', 'renal']]
```

### acidbase

```python
merged_df = pd.merge(bg, icustay, on='hadm_id')
merged_df = merged_df[(merged_df['charttime'] >= merged_df['intime']) & (merged_df['charttime'] < merged_df['outtime']) & (merged_df['specimen_pred'] == 'ART.')]

# Filter out rows where either ph or pco2 is null
filtered_df = merged_df.dropna(subset=['ph', 'pco2']).reset_index()

# Define a function to calculate the acidbase_score
def calculate_acidbase_score(row):
    ph, pco2 = row['ph'], row['pco2']
    if ph < 7.20:
        return 12 if pco2 < 50 else 4
    elif ph < 7.30:
        if pco2 < 30:
            return 9
        elif pco2 < 40:
            return 6
        elif pco2 < 50:
            return 3
        else:
            return 2
    elif ph < 7.35:
        return 9 if pco2 < 30 else (0 if pco2 < 45 else 1)
    elif ph < 7.45:
        return 5 if pco2 < 30 else (0 if pco2 < 45 else 1)
    elif ph < 7.50:
        if pco2 < 30:
            return 5
        elif pco2 < 35:
            return 0
        elif pco2 < 45:
            return 2
        else:
            return 12
    elif ph < 7.60:
        return 3 if pco2 < 40 else 12
    else:  # ph >= 7.60
        return 0 if pco2 < 25 else (3 if pco2 < 40 else 12)

# Apply the function to calculate the acidbase_score
filtered_df['acidbase_score'] = filtered_df.apply(calculate_acidbase_score, axis=1)

```

```python
def calculate_hr_score(row):
    if pd.isnull(row['heart_rate_max']):
        return None
    # Calculate absolute differences
    abs_diff_max = abs(row['heart_rate_max'] - 75)
    abs_diff_min = abs(row['heart_rate_min'] - 75)

    # Implementing the CASE logic
    if abs_diff_max > abs_diff_min:
        return row['hr_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['hr_score_smin']
    elif abs_diff_max == abs_diff_min:
        if row['hr_score_smax'] >= row['hr_score_smin']:
            return row['hr_score_smax']
        else:
            return row['hr_score_smin']
    else:
        return None  # This else is optional as all cases are already covered

# Apply the function to each row
df['hr_score'] = df.apply(calculate_hr_score, axis=1)
```

```python
def calculate_mbp_score(row):
    if pd.isnull(row['mbp_max']):
        return None
    # Calculate absolute differences
    abs_diff_max = abs(row['mbp_max'] - 90)
    abs_diff_min = abs(row['mbp_min'] - 90)

    # Implementing the CASE logic
    if abs_diff_max > abs_diff_min:
        return row['mbp_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['mbp_score_smin']
    elif abs_diff_max == abs_diff_min:
        if row['mbp_score_smax'] >= row['mbp_score_smin']:
            return row['mbp_score_smax']
        else:
            return row['mbp_score_smin']
    else:
        return None  # This else is optional as all cases are already covered

# Apply the function to each row
df['mbp_score'] = df.apply(calculate_mbp_score, axis=1)
```

```python
def calculate_temp_score(row):
    if pd.isnull(row['temperature_max']):
        return None
    abs_diff_max = abs(row['temperature_max'] - 38)
    abs_diff_min = abs(row['temperature_min'] - 38)

    if abs_diff_max > abs_diff_min:
        return row['temp_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['temp_score_smin']
    elif abs_diff_max == abs_diff_min:
        if row['temp_score_smax'] >= row['temp_score_smin']:
            return row['temp_score_smax']
        else:
            return row['temp_score_smin']

# Function to calculate respiratory rate score
def calculate_resp_rate_score(row):
    if pd.isnull(row['resp_rate_max']):
        return None
    abs_diff_max = abs(row['resp_rate_max'] - 19)
    abs_diff_min = abs(row['resp_rate_min'] - 19)

    if abs_diff_max > abs_diff_min:
        return row['resp_rate_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['resp_rate_score_smin']
    elif abs_diff_max == abs_diff_min:  # This condition is always true as per the provided logic, seems like a typo. Assuming it meant to compare with resp_rate_min
        if row['resp_rate_score_smax'] >= row['resp_rate_score_smin']:
            return row['resp_rate_score_smax']
        else:
            return row['resp_rate_score_smin']

# Apply functions to each row
df['temp_score'] = df.apply(calculate_temp_score, axis=1)
df['resp_rate_score'] = df.apply(calculate_resp_rate_score, axis=1)
```

```python
# Function to calculate hematocrit score
def calculate_hematocrit_score(row):
    if pd.isnull(row['hematocrit_max']):
        return None
    abs_diff_max = abs(row['hematocrit_max'] - 45.5)
    abs_diff_min = abs(row['hematocrit_min'] - 45.5)

    if abs_diff_max > abs_diff_min:
        return row['hematocrit_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['hematocrit_score_smin']
    else:  # When abs_diff_max == abs_diff_min
        if row['hematocrit_score_smax'] >= row['hematocrit_score_smin']:
            return row['hematocrit_score_smax']
        else:
            return row['hematocrit_score_smin']

# Function to calculate white blood cell (WBC) score
def calculate_wbc_score(row):
    if pd.isnull(row['wbc_max']):
        return None
    abs_diff_max = abs(row['wbc_max'] - 11.5)
    abs_diff_min = abs(row['wbc_min'] - 11.5)

    if abs_diff_max > abs_diff_min:
        return row['wbc_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['wbc_score_smin']
    else:  # When abs_diff_max == abs_diff_min
        if row['wbc_score_smax'] >= row['wbc_score_smin']:
            return row['wbc_score_smax']
        else:
            return row['wbc_score_smin']

# Apply the functions to each row
df['hematocrit_score'] = df.apply(calculate_hematocrit_score, axis=1)
df['wbc_score'] = df.apply(calculate_wbc_score, axis=1)
```

```python
# Function to calculate creatinine score
def calculate_creatinine_score(row):
    if pd.isnull(row['creatinine_max']):
        return None
    if row['arf'] == 1:
        return row['creatinine_score_smax']

    abs_diff_max = abs(row['creatinine_max'] - 1)
    abs_diff_min = abs(row['creatinine_min'] - 1)

    if abs_diff_max > abs_diff_min:
        return row['creatinine_score_smax']
    elif abs_diff_max < abs_diff_min:
        return row['creatinine_score_smin']
    else:  # When values are equidistant
        if row['creatinine_score_smax'] >= row['creatinine_score_smin']:
            return row['creatinine_score_smax']
        else:
            return row['creatinine_score_smin']

# Apply the function to each row
df['creatinine_score'] = df.apply(calculate_creatinine_score, axis=1)
```

```python
def calculate_bun_score(row):
    if pd.isnull(row['bun_max']):
        return None
    else: return row['bun_score_smax']

df['bun_score'] = df.apply(calculate_bun_score, axis=1)
```

```python
def calculate_sodium_score(row):
    if pd.isnull(row['sodium_max']):
        return None
    max_diff = abs(row['sodium_max'] - 145.5)
    min_diff = abs(row['sodium_min'] - 145.5)
    if max_diff > min_diff:
        return row['sodium_score_smax']
    elif min_diff > max_diff:
        return row['sodium_score_smin']
    else:  # Values are equidistant, pick the larger score
        if row['sodium_score_smax'] >= row['sodium_score_smin']:
            return row['sodium_score_smax']
        else:
            return row['sodium_score_smin']

df['sodium_score'] = df.apply(calculate_sodium_score, axis=1)
```

```python
def calculate_albumin_score(row):
    if pd.isnull(row['albumin_max']):
        return None
    max_diff = abs(row['albumin_max'] - 3.5)
    min_diff = abs(row['albumin_min'] - 3.5)
    if max_diff > min_diff:
        return row['albumin_score_smax']
    elif min_diff > max_diff:
        return row['albumin_score_smin']
    else:  # Values are equidistant, choose the larger score
        if row['albumin_score_smax'] >= row['albumin_score_smin']:
            return row['albumin_score_smax']
        else:
            return row['albumin_score_smin']

# Apply the function to each row
df['albumin_score'] = df.apply(calculate_albumin_score, axis=1)
```

```python
def calculate_glucose_score(row):
    if pd.isnull(row['glucose_max']):
        return np.nan
    elif abs(row['glucose_max'] - 130) > abs(row['glucose_min'] - 130):
        return row['glucose_score_smax']
    elif abs(row['glucose_max'] - 130) < abs(row['glucose_min'] - 130):
        return row['glucose_score_smin']
    elif abs(row['glucose_max'] - 130) == abs(row['glucose_min'] - 130):
        if row['glucose_score_smax'] >= row['glucose_score_smin']:
            return row['glucose_score_smax']
        else:
            return row['glucose_score_smin']

# Apply the function to each row
df['glucose_score'] = df.apply(calculate_glucose_score, axis=1)
```

```python
def calculate_uo_score(row):
    if pd.isnull(row['urineoutput']):
        return None
    elif row['urineoutput'] < 400:
        return 15
    elif row['urineoutput'] < 600:
        return 8
    elif row['urineoutput'] < 900:
        return 7
    elif row['urineoutput'] < 1500:
        return 5
    elif row['urineoutput'] < 2000:
        return 4
    elif row['urineoutput'] < 4000:
        return 0
    elif row['urineoutput'] >= 4000:
        return

# Apply the function to the urineoutput column to calculate the uo_score for each row
df['uo_score'] = df.apply(calculate_uo_score, axis=1)
```

```python
def calculate_gcs_score(row):
    if row['gcs_unable'] == 1:
        return 0  # Assuming normal for intubated patients as per original study
    elif row['gcs_eyes'] == 1:
        if row['gcs_verbal'] == 1 and row['gcs_motor'] in [1, 2]:
            return 48
        elif row['gcs_verbal'] == 1 and row['gcs_motor'] in [3, 4]:
            return 33
        elif row['gcs_verbal'] == 1 and row['gcs_motor'] in [5, 6]:
            return 16
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] in [1, 2]:
            return 29
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] in [3, 4]:
            return 24
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] >= 5:
            return np.nan  # Unlikely combination, returning NaN
        elif row['gcs_verbal'] >= 4:
            return np.nan
    elif row['gcs_eyes'] > 1:
        if row['gcs_verbal'] == 1 and row['gcs_motor'] in [1, 2]:
            return 29
        elif row['gcs_verbal'] == 1 and row['gcs_motor'] in [3, 4]:
            return 24
        elif row['gcs_verbal'] == 1 and row['gcs_motor'] in [5, 6]:
            return 15
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] in [1, 2]:
            return 29
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] in [3, 4]:
            return 24
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] == 5:
            return 13
        elif row['gcs_verbal'] in [2, 3] and row['gcs_motor'] == 6:
            return 10
        elif row['gcs_verbal'] == 4 and row['gcs_motor'] in [1, 2, 3, 4]:
            return 13
        elif row['gcs_verbal'] == 4 and row['gcs_motor'] == 5:
            return 8
        elif row['gcs_verbal'] == 4 and row['gcs_motor'] == 6:
            return 3
        elif row['gcs_verbal'] == 5 and row['gcs_motor'] in [1, 2, 3, 4, 5]:
            return 3
        elif row['gcs_verbal'] == 5 and row['gcs_motor'] == 6:
            return 0
    return np.nan  # For cases that do not match any condition

# Apply the function to each row
df['gcs_score'] = df.apply(calculate_gcs_score, axis=1)
```
